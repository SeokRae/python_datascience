## 결측치 다루기
 - isnull().sum()
 - plot()/plot.bar()/plot.harh()
 - reset_index()
 - sort_values()
 - NaN
 - drop()

```python
# 결측치 확인
# boolean indexing으로 True는 1이므로 isnull()을 통하여 갯수 파악
null_count = df.isnull().sum()
null_count
```

## 결측치 시각화하기
### 기본그래프

```python
# line 그래프
null_count.plot()
```

```python
# bar 그래프
null_count.plot.bar()
```

```python
# 글짜 기울이기
null_count.plot.bar(rot=60)
```

```python
# barh 그래프 (세로 막대형 그래프)
null_count.plot.barh(figsize=(5,7))
```

### 그래프 데이터프레임으로 변경하기

```python
# 인덱스 초기화하기
type(null_count)
```

```python
type(null_count.reset_index())
```

```python
# 데이터프레임화 하기
df_null_count = null_count.reset_index()
df_null_count.head(3)
```

```python
# 컬럼명 변경하기
df_null_count.columns = ["컬럼명", "결측치수"]
df_null_count.head()
```

### 결측치만 보기

```python
# 결측치수 컬럼을 기준으로 상위 10개만 보기 ( 내림차순 )
df_null_count_top = df_null_count.sort_values(by="결측치수", ascending=False).head(10)
df_null_count_top
```

### 특정컬럼만 보기

```python
# 틀정 컬럼만 보기
df["지점명"].head()
```

```python
df["지점명"].tail()
```

### column 명 리스트화 하기

```python
# 결측치데이터 중 "컬럼명" 상위 10개를 가져옴
df_null_count_top["컬럼명"]
```

```python
# 컬럼명 Series를 to List casting
drop_columns = df_null_count_top["컬럼명"].tolist()
drop_columns
```

### columns 삭제

```python
# dataframe에서 columns 삭제하기전 데이터 확인
df[drop_columns].head()
```

```python
# 원본 확인
df
```

```python
# 데이터프레임 컬럼 삭제 전 확인 
df.shape
```

```python
# 데이프레임에서 특정 컬럼 삭제 (axis 1은 열, 0은 행)
df = df.drop(drop_columns, axis=1)
df
```

### 요약 값 확인하기
 - 25%는 앞에서 1/4 되는 값이고, 1사분위수

 - 75%는 앞에서 3/4 번째 되는 수고, 3사분위수
 - mean, min, max는 각각 평균, 최소값, 최대값
